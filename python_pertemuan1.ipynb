1. 5 nama mahasiswa


```python
mahasiswa = ["Andi", "Budi", "Citra", "Dewi", "Eko"]
print("List nama mahasiswa :")
print(mahasiswa)
```

Tambahkan 1 nama menggunakan .append()

```python
mahasiswa.append("dania")
print("\nSetelah menambahkan nama baru:")
print(mahasiswa)
```

hapus elemen menggunakan .remove()

```python
mahasiswa.remove("Citra")
print("\nSetelah menghapus nama Citra:")
print(mahasiswa)
```
